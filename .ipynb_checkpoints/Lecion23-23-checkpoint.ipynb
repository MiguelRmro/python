
# PRÁCTICA 5: ¡JUGUEMOS BOP-IT!
## Bop-it es un juego de Hasbro creado en los años 90 y muy exitoso a comienzos de los 2000.

Es un juego audible lo que quiere decir que el usuario debe seguir los comandos de voz generados por el juego para presionar, girar o halar diferentes elementos del mismo a medida que la velocidad del juego aumenta.

Se recibe un punto cada vez que la acción ejecutada por el jugador coincide con el comando de voz, mientras que el juego termina si la acción es equivocada. La dificultad del juego (es decir la rapidez con que los comandos de voz se presentan al usuario) se incrementa a medida que se incrementan los aciertos.

En las primeras versiones del juego se tenían 3 comandos de voz, equivalentes a tres acciones a ejecutar (ver figura de abajo):

Bop-it: el usuario debe presionar el botón central
Gira: el usuario debe girar el control de color amarillo
Estira: el usuario debe halar el control de color azul


En esta última práctica la idea es implementar un programa en Python que sea capaz de emular el juego Bop-it, pero usando el teclado en lugar de controles mecánicos.

Para hacerlo usaremos varios de los conceptos que hemos visto a lo largo del curso, como la lectura desde el teclado, así como sentencias if y while junto con el concepto de funciones que vimos en esta última parte del curso.

Pero además incorporaremos otras herramientas, como el uso de algunas librerías de Python para agregar aleatoriedad al juego, para reproducir sonidos y para agregar mayor dificultad al juego.

# Requerimientos
## En detalle, el juego a implementar debe cumplir con los siguientes requerimientos (que son tomados del juego original):

Debe manejar tres estados: "bop-it", "gira" y "estira"
A cada estado se asociará un comando de voz con el nombre del estado
En cada ronda el estado (y por tanto el comando de voz) serán presentados al usuario de forma aleatoria
La interacción del usuario se dará a través del teclado de la siguiente manera:
Comando "bop-it": tecla "s" + Enter
Comando "gira": tecla "d" + Enter
Comando "estira": tecla "a" + Enter
En cada ronda el usuario contará con un tiempo pre-definido para presionar la tecla correspondiente. Este tiempo debe disminuir progresivamente a medida que avanza el juego, incrementando así su nivel de dificultad
Si el usuario presiona la tecla incorrecta, o no lo hace en el tiempo establecido, el juego terminará inmediatamente.
A lo largo del juego se debe almacenar y actualizar el puntaje obtenido, y al final del mismo se debe imprimir en pantalla
A pesar de que son muchos los requerimientos veremos que la implementación en Python resultará relativamente sencilla.

# Pseudo-código

```python
# 1. seleccionar_estado (función)
# 2. comando_voz (función)
# 3. interaccion_usuario (función)

# 4. Bloque central "while" que dependa de una variable "success"
# - Selección aleatoria de 1 de 3 estados -> función seleccionar_estado
# - Reproducir el sonido asociado al estado seleccionado -> función comando_voz
# - Iniciar cuenta regresiva:
#   - Leer entrada del teclado del usuario -> función interaccion_usuario
# - Al terminar la cuenta regresiva
#   - Verificar si la entrada coincide o no con el estado presentado -> función interaccion_usuario
#     - Sí: sumar 1 punto. Incrementar dificultad (reducir la duración de la cuenta regresiva)
#     - No: el juego termina
#   - Si termina regresiva y el usuario no presiona ninguna tecla el juego termina
```

# Implementación

```python
# Función 1: seleccionar aleatoriamente uno de los tres estados

import random
# from random import choice
def seleccionar_estado():
    '''Seleccionar aleatoriamente uno de los tres posibles estados'''
    estado = random.choice(("gira","estira","bop-it"))
    return estado    
```

```python
estado = seleccionar_estado()
print(estado)
```

```python
# Función 2: reproducir comando de voz

from IPython.display import Audio, display

def comando_voz(estado):
    '''Reproduce el comando de voz asociado al estado seleccionado'''
    
    # Leer el archivo de audio
    sonido = Audio(filename='./'+estado+'.wav', autoplay=True)
    
    # Reproducir el sonido
    display(sonido, clear=True)
```

```python
# Función 2: reproducir comando de voz

from IPython.display import Audio, display

def comando_voz(estado):
    '''Reproduce el comando de voz asociado al estado seleccionado'''
    
    # Leer el archivo de audio
    sonido = Audio(filename='./'+estado+'.wav', autoplay=True)
    
    # Reproducir el sonido
    display(sonido, clear=True)
```

```python
comando_voz(estado)
```

```python
# Función 3: presionar tecla y verificar si coincide o no con el estado

def interaccion_usuario(estado):
    '''Solicita al usuario presionar la tecla y verifica si esta coincide con el estado'''
    
    # Solicitar entrada al usuario
    tecla = input('')
    
    # Convertir la tecla oprimida a minúsculas en caso de que el teclado esté en mayúscula
    tecla = tecla.lower()
    
    # Verificación
    success = False
    
    if (estado=='bop-it') and (tecla=='s'):
        success = True
    if (estado=='gira') and (tecla=='d'):
        success = True
    if (estado=='estira') and (tecla=='a'):
        success = True
    
    return success
```

```python
# Prueba de las tres funciones
estado = seleccionar_estado()
comando_voz(estado)
success=interaccion_usuario(estado)
```

```python
# Bloque principal (while)

from threading import Timer # Cuenta regresiva

# Inicializar variables
puntaje = 0
timeout = 3 # El tiempo asignado inicialmente al usuario
delta = 0.5 # El tiempo (en segundos) que se irá reduciendo la cuenta regresiva
success = True

while success:
    # Seleccionar el estado aleatoriamente
    estado = seleccionar_estado()
    
    # Crear el timer, iniciarlo, reproducir comando de voz e interacción del usuario
    t = Timer(timeout, print, '⏰ ⏰ ⏰ SE ACABO EL TIEMPO ⏰ ⏰ ⏰')
    t.start() # Iniciar la cuenta regresiva
    comando_voz(estado) # Reproducir estado como comando de voz
    success=interaccion_usuario(estado) # Verificar interacción del usuario
    t.cancel() # Terminar la cuenta regresiva
    
    if t.is_alive() == False: # Si terminó la cuenta regresiva y no hubo interacción
        break
    
    # Si el usuario responde correctamente sumar puntaje y reducir tiempo
    if success:
        puntaje += 1
        if timeout - delta > 0: # Evitar valores negativos 
            timeout -= delta

print(f'🏆 🏆 🏆 PUNTAJE FINAL: {puntaje} puntos 🏆 🏆 🏆')
    
```
